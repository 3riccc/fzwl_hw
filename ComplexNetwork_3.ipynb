### 3.1生成小世界网络

调用函数：
 nx.random_graph.watts_strongz_graph(n,k,p)
- 网络规模 n
- 最近邻居数 k
- 断边重连概率 p

````python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
````

````python
#生成小世界网络

N = 30 # 节点数
m = 4 # 起始连接的最近邻数
p = 0.1 # 断边重连概率

g = nx.watts_strogatz_graph(N, m, p)
f=plt.figure(figsize=(5,4))
nx.draw_shell(g,node_size=200,with_labels=True, font_weight='bold')
plt.show()
````

````python
#画图的技巧
fig,ax=plt.subplots(figsize=(8,6))
layout=[nx.shell_layout,
        nx.layout.circular_layout,
        nx.layout.fruchterman_reingold_layout,
        nx.layout.circular_layout,
        nx.layout.kamada_kawai_layout,
        nx.layout.spring_layout]

pos=layout[4](g)

NodeId=list(g.nodes())
node_color = NodeId
node_size = [g.degree(i)**2.5*25 for i in NodeId]
options = {
    'node_size': node_size,
    'line_color': 'grey',
    'linewidths': 0.1,
    'width': 0.4,
    'style':'solid',
    'nodelist':NodeId,
    'node_color':node_color,
    'font_color':'w',

}
nx.draw(g,pos=pos,ax=ax,with_labels=True,**options)
plt.show()
````

### 计算网络的平均最短路径和集聚系数
#### 集聚系数

$$c_i=\frac{\sum_j\sum_m A_{ij}A_{jm}A_{mi}}{k_i(k_i-1)}$$

$$C = \langle c_i \rangle$$

```python
nx.clustering(G)   #返回所有节点的集聚系数
nx.clustering(G,1) #返回特定节点的集聚系数
```

#### 最短距离

节点间的最短距离$d_{ij}$表示从节点i最少走多少距离可以到节点j。

```python
nx.shortest_path_length(G,source=1,target=2) #返回特定节点间的最短距离
nx.shortest_path_length(G,source=1) #返回特定节点与其它所有节点的最短距离
nx.shortest_path_length(G) #返回所有节点间的最短距离
```
### 最近邻居网络的平均最短路径和平局集聚系数

````python
N = 1000
m = 4

g = nx.watts_strogatz_graph(N,m,0)
d0 = nx.average_shortest_path_length(g)
c0 = nx.average_clustering(g)

print('平均最短路径： %.3f'%d0)
print('平均集聚系数： %.3f'%c0)
````

### 小世界特性曲线

````python
p = 1.0
pc = 0.6
prange = []
drange = []
crange = []
for i in range(20):
    prange.append(p)
    d = []
    c = []
    for j in range(5):
        print(i,j,end='\r')
        g = nx.watts_strogatz_graph(N,m,p)
        d.append(nx.average_shortest_path_length(g))
        c.append(nx.average_clustering(g))
    d = np.mean(d)
    c = np.mean(c)
    drange.append(d/d0)
    crange.append(c/c0)
    p *= pc
````

````python
f,(ax1,ax2)=plt.subplots(figsize=(8,4),nrows=1,ncols=2)

options1={
    "markeredgecolor":'g',
    "markerfacecolor":'g',
    "markersize":8,
    "marker":"o",
    "label":'spl',
    "alpha":0.8,
    "linestyle":' ',
       
}
ax1.plot(prange, drange,**options1)

options2={
    "markeredgecolor":'b',
    "markerfacecolor":None,
    "markersize":8,
    "marker":"s",
    "label":'cls',
    "alpha":0.8,
    "linestyle":' ',
       
}

ax1.plot(prange, crange, **options2)
ax1.set_xlim(2e-5,2)
ax1.set_ylim(0,1.05)
ax1.set_xscale('log')
ax1.legend(loc='lower left',numpoints=1)
ax1.set_xlabel('p', fontsize=14)


ratio = [crange[i]/drange[i] for i in range(len(crange))]

options3={
    "markerfacecolor":'orangered',
    "marker":"o",
    "label":r'$\frac{C}{d}$',
    "alpha":0.8,
    "linestyle":' ',
       
}
ax2.plot(prange, ratio,**options3)
ax2.set_xlim(2e-5,2)
ax2.set_xscale('log')
ax2.legend(loc='upper left',numpoints=1)
ax2.set_xlabel('p', fontsize=14)
plt.show()
````
